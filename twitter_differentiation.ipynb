## First step is to collect the data from different twitter accounts

```python
#Getting Tweets
import tweepy
import pickle

from twitter_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_secret)

api = tweepy.API(auth)


public_tweets=api.user_timeline('@realDonaldTrump')

with open('donaldtrump.pkl', 'wb') as f:
    pickle.dump(public_tweets,f)
    
    
public_tweets2=api.user_timeline('@jimmyfallon')

with open('jimmyfallon.pkl', 'wb') as f:
    pickle.dump(public_tweets2,f)
    
public_tweets3=api.user_timeline('@BearGrylls')

with open('beargrylls.pkl', 'wb') as f:
    pickle.dump(public_tweets3,f)
    

```
